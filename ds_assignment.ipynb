```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
data =pd.read_csv("C:/Users/arjun/Downloads/ABC Company.csv")
data.head()
```

```python
## here the columns height and weight has been reworked to ensure data consistency
data['Height']= np.random.randint(150,181,size=len(data))
data['Weight']=np.random.randint(50,101,size=len(data))
data.head()
data_clean=data.dropna()
```

```python
##distribution across each team
plt.figure(figsize=(18,18))
employee_count=data_clean['Team'].value_counts().sort_values(ascending=True)
print(employee_count)
plt.barh(employee_count.index,employee_count,color='lightblue')
plt.title("Distribution of employees",size=30,loc='center')
plt.xlabel('Count',size=20)
plt.ylabel('Team Name',size=20)
plt.xticks(size=16)
plt.yticks(size=16)
plt.show()

##team New Orleans Pelicans has the highest number of members and Minnesota Timberwolves with the lowest in number. only 5 teams consist of members less than or equal to 10 in count.
```

```python
##percentage split of employees
plt.figure(figsize=(15,15))
percentage= (employee_count/employee_count.sum())*100
print(percentage)
plt.pie(employee_count,autopct='%1.1f%%',labels=employee_count.index)
plt.title("Percentage distribution of employees",size=20)
## pie chart depicts almost an even distribution of employees over different teams.
```

```python
##segragation based on role
d1=data_clean.groupby('Position').size()
plt.figure(figsize=(10,10))
plt.bar(d1.index,d1)
plt.title("Distribution of Employees over various Positions",size=20)
plt.xlabel("Positions",size=16)
plt.ylabel("No. of Employees",size=16)
plt.show()
## employees with position SG dominates the organisation, whereas the employees with position C have a smaller share of workforce.
```

```python
##predominant age group
bins = [20, 30, 40, 50, 60]
labels = ['20-29', '30-39', '40-49', '50-59']

data_clean['AgeGroup'] = pd.cut(data_clean['Age'], bins=bins, labels=labels, right=False)

age_group_count = data_clean.groupby('AgeGroup').size()
print(age_group_count)

predominant_group = age_group_count.idxmax()
print("Predominant Age Group:", predominant_group)

plt.figure(figsize=(8,6))
plt.bar(age_group_count.index, age_group_count.values, color='skyblue')
plt.title("Employee Distribution by Age Group", fontsize=14)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.show()
##employees under age 30 is high in number compared to other age groups.
```

```python
## team and position with highest salary expenditure

team_salary = data_clean.groupby('Team')['Salary'].sum().sort_values(ascending=False)
print("Team-wise salary expenditure:")
print(team_salary)

highest_team = team_salary.idxmax()
highest_team_value = team_salary.max()
print(f"\nTeam with highest salary expenditure: {highest_team} - {highest_team_value}")

position_salary = data_clean.groupby('Position')['Salary'].sum().sort_values(ascending=False)
print("\nPosition-wise salary expenditure:")
print(position_salary)

highest_position = position_salary.idxmax()
highest_position_value = position_salary.max()
print(f"\nPosition with highest salary expenditure: {highest_position} - {highest_position_value}")

plt.figure(figsize=(10,5))
team_salary.plot(kind='bar', title='Salary Expenditure by Team')
plt.ylabel('Total Salary')
plt.xlabel('Team')
plt.show()
plt.figure(figsize=(10,5))
position_salary.plot(kind='bar', color='orange', title='Salary Expenditure by Position')
plt.ylabel('Total Salary')
plt.xlabel('Position')
plt.show()
## Team Los Angeles Clippers holds the highest share of team expenditure and PG bags first in terms of expenditure based on position.
```

```python
##correlation b/w age and salary
data_clean.loc['Age'] = pd.to_numeric(data_clean['Age'], errors='coerce')
data_clean.loc['Salary'] = pd.to_numeric(data_clean['Salary'], errors='coerce')
correlation = data_clean['Age'].corr(data_clean['Salary'])
print(f"Correlation b/w Age and Salary: {correlation:.2f}")

plt.figure(figsize=(8,6))
sns.scatterplot(x='Age', y='Salary', data=data_clean, color='blue', s=50)
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()

## the result obatined is a positive value which implies that it is positive correlation.
```
